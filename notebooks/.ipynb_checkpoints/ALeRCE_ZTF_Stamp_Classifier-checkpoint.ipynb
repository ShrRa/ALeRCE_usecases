### Describe functions to get light curves

```python
def plotLC(oid, SN_det, SN_nondet):
    
    fig, ax = plt.subplots(figsize = (14, 7))
    labels = {1: 'g', 2: 'r'}
    colors = {1: 'g', 2: 'r'}
    markers = {1: 'o', 2: 's'}
    sizes = {1: 30, 2: 60}
    
    # loop the passbands
    for fid in [1, 2]:
        
        # plot detections if available
        mask = SN_det.fid == fid
        if np.sum(mask) > 0:
            # note that the detections index is candid and that we are plotting the psf corrected magnitudes
            ax.errorbar(SN_det[mask].mjd, SN_det[mask].magpsf, 
                yerr = SN_det[mask].sigmapsf, c=colors[fid], label=labels[fid], marker=markers[fid])
        
        # plot non detections if available
        mask = (SN_nondet.fid == fid) & (SN_nondet.diffmaglim > -900)
        if np.sum(mask) > 0:     
            # non detections index is mjd
            ax.scatter(SN_nondet[mask].index, SN_nondet[mask].diffmaglim, c=colors[fid], alpha = 0.5,
                marker='v', label="lim.mag. %s" % labels[fid], s=sizes[fid])
            
    ax.set_title(oid)
    ax.set_xlabel("MJD")
    ax.set_ylabel("Apparent magnitude")
    ax.legend()
    ax.set_ylim(ax.get_ylim()[::-1])
    

def getSNdata(oid, doLC=False, dostamp=False):

    results = {"oid": oid}
        
    # query detections
    SN_det = client.get_detections(oid, format='pandas')
    results["lc_det"] = SN_det
        
    # query non detections
    SN_nondet = client.get_non_detections(oid, format='pandas')
    results["lc_nondet"] = SN_nondet
    
    # plot the LC
    if doLC:
        plotLC(oid, SN_det, SN_nondet)
        
    # show the first image stamp
    if dostamp:
        client.plot_stamp(oid)    
        
    # return data
    return results
```

### Describe how to get objects per class

```python
def get_objects_per_class(classearly="sn", pclassearly=0.5, min_nobs=10, n_objects=100):

    total = n_objects
    records_per_page = n_objects
    page = 1
    sortBy = "pclassearly"
    max_nobs = 50
    min_firstmjd = Time("2017-11-01T00:00:00", format="isot", scale="utc").mjd + 70

    params = {
        #"total": total,
        "records_per_pages": records_per_page,
        "page": page,
        "sortBy": sortBy,
        "query_parameters": {
            "filters": {
                "pclassearly": pclassearly,
                "classearly": classearly,
                "nobs": {
                    "min": min_nobs,
                    "max": max_nobs
                },
                "dates": {
                    "firstmjd": {
                        "min": min_firstmjd
                    }
                }
            }
        }
    }
    
    objects = client.query(params, format='pandas')
    print(objects.shape)
    
    return objects
```

```python
def plot_some_LC(df, n_examples=3):
    for i in range(n_examples):
        object_id = df.iloc[i].name
        print("https://alerce.online/object/"+object_id)
        getSNdata(object_id, doLC=True, dostamp=True)
        plt.show()
```

```python
n_objects = 40000
early_classes = {"agn": 18, "sn": 19, "vs":20, "asteroid": 21, "bogus":22}
objects = {}
for cl, num in early_classes.items():
    if cl in ["agn", "sn", "vs"]:
        min_nobs=3
    else:
        min_nobs=1
    objects[cl] = get_objects_per_class(classearly=num, min_nobs=min_nobs, n_objects=n_objects)
```

# Plot some examples per class <a class="anchor" id="examples"></a>

```python
for cl, df in objects.items():
    print("######### "+ cl + " class #########")
    plot_some_LC(df, n_examples=2)
```

### Compute galactic and ecliptic coordinates

```python
def ecliptic_coordinates(df):
    ecl = df.apply(lambda row: ephem.Ecliptic(ephem.Equatorial('%s' % (row.meanra / 15.),
                                                               '%s' % row.meandec, epoch=ephem.J2000)), axis=1)
    df["ecl_lat"] = ecl.apply(lambda row: np.rad2deg(row.lat))
    df["ecl_long"] = ecl.apply(lambda row: np.rad2deg(row.long))
    return df


def galactic_coordinates(df):
    gal = df.apply(lambda row: ephem.Galactic(ephem.Equatorial('%s' % (row.meanra / 15.),
                                                               '%s' % row.meandec, epoch=ephem.J2000)), axis=1)
    df["gal_lat"] = gal.apply(lambda row: np.rad2deg(row.lat))
    df["gal_long"] = gal.apply(lambda row: np.rad2deg(row.long))
    return df

for cl, df in objects.items():
    ecl_df = ecliptic_coordinates(df)
    gal_df = galactic_coordinates(ecl_df)
    objects[cl] = gal_df
```

```python
def plot_spatial_distribution(df, ax, index=0, cmin_val=0.1, vmax_val=5, titles="", fontsize=18):
    dims = ['gal_long', 'gal_lat']
    _, _, _, im = ax.hist2d(df[dims].values[:, 0], df[dims].values[:, 1],
                            (300, 300), cmap="viridis", vmax=vmax_val, cmin=cmin_val)
    ax.tick_params(axis='both', labelsize=fontsize-2)
    cbar = plt.colorbar(im,ax=ax, pad=0)
    cbar.ax.tick_params(labelsize=fontsize-4)
    
    ecliptic_lat = np.linspace(0, 0, num=500)
    ecliptic_longi = np.linspace(0, 360, num=500)
    galact_long, galact_lat = ec2gal(ecliptic_longi, ecliptic_lat)
    
    ax.set_title(titles, fontsize=fontsize)
    ax.plot(galact_long, galact_lat, "-k", lw=6)
    ax.plot(galact_long, galact_lat, "-y", lw=4)
    
    return ax
```

```python
f, ax = plt.subplots(2, 3, figsize=(3*10, 2*6))
plt.subplots_adjust(wspace=0.0, hspace=0.15)
ax = ax.flatten()
titles = ["all objects", ] + list(objects.keys())

for i in range(6):
    if i == 0:
        df = pd.concat(list(objects.values()), axis=0)
    else:
        df = list(objects.values())[i-1]
    
    ax[i] = plot_spatial_distribution(df, ax[i], titles=titles[i])
plt.show()
```

### Query objects uniformly in over full ZTF sky coverage to make de distribution invariant to the density

```python
def get_uniform_in_space():

    min_firstmjd = Time("2017-11-01T00:00:00", format="isot", scale="utc").mjd + 70
    query_params = get_space_uniform_query(min_firstmjd, n_samples_per_cone=2500)
    object_list = []
    for params in tqdm(query_params, disable=False):
        try:
            objects = client.query(params, format="pandas")
            if objects is None:
                continue
        except:
            continue
        objects.fillna(value=np.nan)
        object_list.append(objects)
    object_list = pd.concat(object_list, axis=0)
    object_list.drop_duplicates(inplace=True)
    return object_list
    
def get_space_uniform_query(min_firstmjd, n_samples_per_cone=100, ra_range=[0, 360], dec_range=[-90, 90],
                            ra_step=10, dec_step=10, search_radius=10.0, complete_map=True, 
                            single_det_per_cone=500):

    ra_list = np.arange(ra_range[0], ra_range[1], step=ra_step).astype(np.float)
    dec_list = np.arange(dec_range[0], dec_range[1], step=dec_step).astype(np.float)
    approx_queried_alerts = len(ra_list)*len(dec_list)*n_samples_per_cone

    query_list = []
    if complete_map:
        effective_radius = search_radius*np.sqrt(2)
    else:
        effective_radius = search_radius

    for ra in ra_list:
        for dec in dec_list:
            params = {"query_parameters": {
                        "filters": {
                        },
                             "firstmjd": {
                                "min": min_firstmjd},

                        "coordinates": {
                            "ra": ra,
                            "dec": dec,
                            "sr": effective_radius*3600},
                        },
                        "page": 1,
                        "records_per_pages": n_samples_per_cone,
                        "total": n_samples_per_cone,
                        "sortBy": "nobs",
                        "sortDesc": True}

            single_det_params = copy.deepcopy(params)
            single_det_params["query_parameters"]["filters"] = {"nobs": {"min": 1, "max": 1}}
            single_det_params["sortBy"] = "lastmjd"
            single_det_params["total"] = single_det_per_cone
            single_det_params["records_per_pages"] = single_det_per_cone

            query_list.append(params)
            query_list.append(single_det_params)

    return query_list
```
